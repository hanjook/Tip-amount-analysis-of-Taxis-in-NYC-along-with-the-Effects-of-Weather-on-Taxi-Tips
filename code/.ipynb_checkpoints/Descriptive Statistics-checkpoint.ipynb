```python
# code from lab 2

from numpy import log, log2

def sturges(x):
    return int(log2(x)) + 1

def rice(x):
    return int(2 * x ** (1/3))

def scott(large, small, sd, x):
    return int((large - small) / (3.5 * (sd/x ** (1/3))))

def freedman(large, small, iqr, x):
    return int((large - small) / (2 * (iqr/x ** (1/3))))
    
def square(x):
    return int(sqrt(x))

def logify(x):
    return log(x) if x else 0
```

```python
# Comparing tip amounts to certain features:

sns.catplot(x = 'passenger_count', y = 'tip_amount', data = dfy, kind = 'bar', color = 'blue')
plt.title('Average Tip amount per Passenger Count')
plt.savefig('../plots/vs_passenger.png')
plt.show()

sns.catplot(x = 'day', y = 'tip_amount', data = dfy, kind = 'bar', color = 'blue')
plt.title('Average Tip amount per Day')
plt.savefig('../plots/vs_day.png')
plt.show()

sns.catplot(x = 'hour', y = 'tip_amount', data = dfy, kind = 'bar', color = 'blue')
plt.title('Average Tip amount per Hour')
plt.savefig('../plots/vs_hour.png')
plt.show()

sns.catplot(x = 'payment_type', y = 'tip_amount', data = dfy, kind = 'bar', color = 'blue')
plt.title('Average Tip amount per Payment Method')
plt.savefig('../plots/vs_payment.png')
plt.show()

sns.catplot(x = 'RatecodeID', y = 'tip_amount', data = dfy, kind = 'bar', color = 'blue')
plt.title('Average Tip amount per Kind of Fare Rate')
plt.savefig('../plots/vs_fare_type.png')
plt.show()
```
