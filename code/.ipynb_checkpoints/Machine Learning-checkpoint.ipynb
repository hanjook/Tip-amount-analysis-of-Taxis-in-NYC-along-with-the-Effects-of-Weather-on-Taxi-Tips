```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dfy = pd.read_feather('../preprocessed_data/final_yellow-2019.feather').dropna()
```

```python
dfy.head()
```

```python
#dfy = dfy.drop(['index', 'level_0'], axis=1)
dfy = dfy.drop('index', axis=1)
```

```python
dfy.tail()
```

```python
dfy.dtypes
```

```python
# Using the features with positive correlation with tip_amount, with 
# code from Lab 3 (in refences)
import statsmodels.formula.api as smf 
fit = smf.ols(formula="tip_amount ~ trip_distance + fare_amount + tolls_amount + total_amount + trip_duration + average_speed",
         data=dfy).fit()
```

```python
fit.summary()
```

```python
from scipy.stats import zscore

yCOLS = ['tip_amount']
xCOLS = ['trip_distance', 'fare_amount', 'tolls_amount', 'total_amount', 'trip_duration', 
         'average_speed', 'day', 'hour', 'RatecodeID', 'passenger_count', 'precipitation', 'avg_temp']

X_train = dfy[xCOLS]
y_train = dfy[yCOLS]

# Categorical and continuous columns
cat_columns=['day', 'hour', 'RatecodeID', 'passenger_count', 'precipitation', 'avg_temp']
con_columns=['trip_distance', 'fare_amount', 'tolls_amount', 'total_amount', 'trip_duration', 
         'average_speed']

# encode the categorical data:
X_train_standard= pd.get_dummies(X_train, columns=cat_columns)

# code from Lab 3 (in refences)
# standardize the continuous data:
X_train_standard[con_columns] = X_train_standard[con_columns].astype(float).apply(zscore)
```

```python
y_train.shape
```

```python
test = pd.read_feather('../preprocessed_data/final_yellow-2020.feather')
y_test = test[yCOLS]
X_test = test[xCOLS]

# encode the categorical data:
X_test_standard= pd.get_dummies(X_test, columns=cat_columns)
# standardize the continuous data:
X_test_standard[con_columns] = X_test_standard[con_columns].astype(float).apply(zscore)
```

```python
print(y_test.shape, X_test_standard.shape)
X_test_standard.tail()
```

```python
X_train_standard.tail()
```

```python
# Fit Multiple Linear Regression Model

from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

#X_train, X_test, y_train,  y_test = train_test_split(X_data, y_data)

model = LinearRegression().fit(X_train_standard, y_train)
```

```python
from sklearn.metrics import mean_squared_error
import math

predictions = model.predict(X_test_standard)
print('RMSE value: ', math.sqrt(mean_squared_error(y_test, predictions)))
print('Accuracy Score: ', model.score(X_test_standard, y_test))
```

```python
# Fit Ridge Regression Model

from sklearn.linear_model import Ridge
from sklearn.metrics import accuracy_score

ridge_fit = Ridge(alpha = 1).fit(X_train_standard, y_train)
y_pred = ridge_fit.predict(X_test_standard)
print('RMSE value: ', math.sqrt(mean_squared_error(y_test, y_pred)))
print('Accuracy Score', ridge_fit.score(X_test_standard, y_test))
```

```python
# Fit Lasso Regression Model

from sklearn.linear_model import Lasso
lasso_fit = Lasso(alpha = 1).fit(X_train_standard, y_train)
y_pred_Lasso = lasso_fit.predict(X_test_standard)
print('RMSE value: ', math.sqrt(mean_squared_error(y_test, y_pred_Lasso)))
print('Accuracy Score', lasso_fit.score(X_test_standard, y_test))
```

```python
# Fit Elastic Net Regression Model

from sklearn.linear_model import ElasticNet
en_fit = ElasticNet(alpha = 1).fit(X_train_standard, y_train)
y_pred_en = en_fit.predict(X_test_standard)
print('RMSE value: ', math.sqrt(mean_squared_error(y_test, y_pred_en)))
print('Accuracy Score', en_fit.score(X_test_standard, y_test))
```
